Feature engineering is where you convert insights → model power.

You already KNOW:

What causes default
Which features matter

Now we encode that knowledge into new features 👇

🚀 STEP 1 — Why Feature Engineering?

Models don’t “think” like humans.

You help the model by creating:

👉 better signals from raw data

🧠 Example:

Instead of:

loan_amnt
annual_inc

👉 Create:

loan_to_income → MUCH stronger feature
🔥 STEP 2 — Create Powerful Features
💰 1. Loan-to-Income Ratio (MOST IMPORTANT)

```python
df['loan_to_income'] = df['loan_amnt'] / df['annual_inc']
```

Meaning:

How big is the loan compared to income?

💥 Strong predictor of default

💳 2. Installment-to-Income Ratio

```python
df['installment_to_income'] = df['installment'] / df['annual_inc']
```

🧠 Meaning:

Monthly burden relative to income

⚠️ 3. Credit Stress Score

```python
df['credit_stress'] = df['dti'] * df['revol_util']
```

🧠 Meaning:

Combined financial pressure

📊 4. Income Buckets

```python
df['income_bucket'] = pd.qcut(df['annual_inc'], q=4)
```

🧠 Meaning:
Low / Medium / High income groups
📊 5. DTI Buckets

```python
df['dti_bucket'] = pd.cut(df['dti'], bins=[0,10,20,30,100])
```

```python
df['int_rate_bucket'] = pd.cut(df['int_rate'], bins=[0,10,15,20,30])
```

🧠 Why Bucketing Helps

Models can learn:

“High DTI group = risky”
instead of exact numbers
🔥 STEP 3 — Encode Categorical Variables
🔤 Convert Categories → Numbers


```python
df = pd.get_dummies(df, drop_first=True)
```

🧠 Why?

ML models only understand numbers

⚠️ Important:

Before encoding:

Remove useless columns like:
emp_title
IDs
🚀 STEP 4 — Feature Selection (Light)

Drop:

Redundant

Keep:

All important + engineered features
