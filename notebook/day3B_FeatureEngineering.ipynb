Feature engineering is where you convert insights → model power.

You already KNOW:

What causes default
Which features matter

Now we encode that knowledge into new features 👇

🚀 STEP 1 — Why Feature Engineering?

Models don’t “think” like humans.

You help the model by creating:

👉 better signals from raw data

🧠 Example:

Instead of:

loan_amnt
annual_inc

👉 Create:

loan_to_income → MUCH stronger feature
🔥 STEP 2 — Create Powerful Features
💰 1. Loan-to-Income Ratio (MOST IMPORTANT)

```python
df['loan_to_income'] = df['loan_amnt'] / df['annual_inc']
```

Meaning:

How big is the loan compared to income?

💥 Strong predictor of default

💳 2. Installment-to-Income Ratio

```python
df['installment_to_income'] = df['installment'] / df['annual_inc']
```

🧠 Meaning:

Monthly burden relative to income

⚠️ 3. Credit Stress Score

```python
df['credit_stress'] = df['dti'] * df['revol_util']
```

🧠 Meaning:

Combined financial pressure

📊 4. Income Buckets

```python
df['income_bucket'] = pd.qcut(df['annual_inc'], q=4)
```

```python
# pd.qcut(df['annual_inc'], q=4)
```

🧠 Meaning:
Low / Medium / High income groups
📊 5. DTI Buckets

```python
df['dti_bucket'] = pd.cut(df['dti'], bins=[0,10,20,30,100])
```

```python
df['int_rate_bucket'] = pd.cut(df['int_rate'], bins=[0,10,15,20,30])
```

🧠 Why Bucketing Helps

Models can learn:

“High DTI group = risky”
instead of exact numbers

```python
# 🧠 Golden Rule Before Encoding

# 👉 Keep columns that:

# Add predictive value
# Are available at loan approval time

# 👉 Drop columns that:

# Cause data leakage
# Are too messy / high-cardinality
# Are post-loan information
# 🚨 DROP THESE (VERY IMPORTANT)

# These will hurt your model or leak future info 👇

# ❌ 1. Direct Leakage Columns
# loan_status ❌ (you already created target from this)

# 👉 Drop immediately

# ❌ 2. Post-Loan Information (BIG NO)

# These are created after loan is issued:

# last_credit_pull_d ❌
# issue_d ❌
# pymnt_plan ❌
# debt_settlement_flag ❌
# hardship_flag ❌

# 👉 These leak future behavior

# ❌ 3. High Cardinality / Useless Text
# emp_title ❌ (too many unique values)

# 👉 Adds noise, not value

# ❌ 4. Redundant / Already Processed
# earliest_cr_line ❌ (you already converted to credit_age)
# ⚠️ CONSIDER DROPPING (OPTIONAL BUT RECOMMENDED)
# ⚠️ sub_grade
# Too granular
# Already covered by grade

# 👉 Keep ONE:

# ✅ Keep grade
# ❌ Drop sub_grade
# ✅ KEEP THESE (IMPORTANT FEATURES)

# These are valuable 👇

# 🏠 Stability
# home_ownership ✅
# 🧾 Risk Assessment
# grade ✅
# verification_status ✅
# 💳 Loan Intent
# purpose ✅
# 🌍 Location (Optional but useful)
# addr_state ✅
# 🏦 Loan Setup
# initial_list_status ✅
# application_type ✅
# disbursement_method ✅
# ⚠️ Special Case
# issue_d

# 👉 If you want:

# Extract year/month → keep transformed version
# Otherwise → drop
```
