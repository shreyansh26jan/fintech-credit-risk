| Status          | Meaning                      |
| --------------- | ---------------------------- |
| Fully Paid      | Loan repaid successfully     |

✅ Keep ONLY:

Charged Off → Bad (1)

❌ Remove:

Late (uncertain)
In Grace Period (temporary)

```python
df = df[df['loan_status'].isin(['Fully Paid', 'Charged Off'])]
```

```python
df['target'] = df['loan_status'].apply(lambda x: 1 if x == 'Charged Off' else 0)
```

Next Step: Feature Selection (Keep vs Drop)

Core Principle (REMEMBER THIS)

👉 Use only pre-loan information

❌ Drop anything that happens after loan is given

```python
# Step 1 — Identify Leakage Columns (DROP ❌)

# These columns leak future information:

# ❌ Drop immediately:
cols_to_drop = [
    'funded_amnt',
    'funded_amnt_inv',
    'total_pymnt',
    'total_pymnt_inv',
    'total_rec_prncp',
    'total_rec_int',
    'total_rec_late_fee',
    'recoveries',
    'collection_recovery_fee',
    'last_pymnt_d',
    'last_pymnt_amnt'
]

df.drop(columns=cols_to_drop, inplace=True, errors='ignore')
# 🧠 Why drop?

# 👉 Example:

# total_pymnt = total money received
# If loan defaulted → low payment

# 👉 Using this = cheating ❌ (data leakage)
```

```python
# Step 2 — Drop Irrelevant Columns

# These don’t help prediction:

cols_to_drop += [
    'id',
    'member_id',
    'url',
    'zip_code',
    'title',
    'policy_code'
]

df.drop(columns=cols_to_drop, inplace=True, errors='ignore')
```

```python
# Step 3 — Drop High Missing Columns

# Run:

missing = df.isnull().mean().sort_values(ascending=False)
missing.head(20)

```

```python
# 👉 Then drop columns with >50% missing:

cols_to_drop = missing[missing > 0.5].index
df.drop(columns=cols_to_drop, inplace=True)
```
